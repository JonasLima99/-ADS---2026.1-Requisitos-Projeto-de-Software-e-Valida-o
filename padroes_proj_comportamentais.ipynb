Atividade Assíncrona 3 - Padrões de Projeto Comportamentais

Chain of Responsibility

Problemas: Evita o acoplamento entre quem envia uma solicitação e quem a recebe. Útil quando vários objetos podem tratar uma requisição e você não quer fixar qual deles fará isso.

Solução: Passa a solicitação por uma "corrente" de objetos. Cada um decide se processa a tarefa ou a empurra para o próximo da fila.

Componentes: Handler (interface para tratar pedidos), ConcreteHandler (trata o pedido ou repassa) e o Client (quem inicia).

Command

Problemas: Quando você precisa parametrizar objetos com ações, enfileirar tarefas ou oferecer a opção de "desfazer" (undo).

Solução: Transforma uma solicitação em um objeto próprio, separando quem solicita a ação de quem realmente a executa.

Componentes: Command (interface com o método execute), ConcreteCommand (liga o Receptor à ação), Invoker (pede a execução) e Receiver (quem faz o trabalho pesado).

Iterator

Problemas: Percorrer elementos de uma coleção (lista, árvore, etc.) sem expor como ela é organizada internamente.

Solução: Move o comportamento de navegação para um objeto separado (o iterador).

Componentes: Iterator (interface com next/hasNext), ConcreteIterator (a lógica de travessia) e a Collection (a estrutura de dados em si).

Mediator

Problemas: Dependências confusas entre muitos objetos (comunicação "todos com todos") que tornam o código difícil de alterar.

Solução: Obriga os objetos a se comunicarem apenas através de um mediador central, centralizando o controle.

Componentes: Mediator (interface de comunicação), ConcreteMediator (coordena os colegas) e Colleagues (os componentes que conversam via mediador).

Memento

Problemas: Necessidade de salvar e restaurar o estado de um objeto sem quebrar as regras de encapsulamento.

Solução: O objeto cria um "retrato" (snapshot) do seu estado interno e o armazena em um objeto externo que não mexe no conteúdo.

Componentes: Originator (objeto que será salvo), Memento (o snapshot) e Caretaker (quem guarda o histórico).

Observer

Problemas: Quando a mudança em um objeto precisa refletir automaticamente em vários outros, mas você não quer que eles fiquem "grudados" (acoplados).

Solução: O objeto principal mantém uma lista de interessados e dispara um aviso geral sempre que algo muda.

Componentes: Subject (o dono da informação) e Observer (quem fica vigiando as mudanças).

State

Problemas: Objetos que mudam drasticamente de comportamento dependendo do seu estado atual (ex: um pedido que pode estar 'Aberto', 'Pago' ou 'Cancelado').

Solução: Em vez de encher o código de if/else, cada estado vira uma classe própria com sua lógica específica.

Componentes: Context (mantém o estado atual) e State (interface comum para os comportamentos de cada estado).

Strategy

Problemas: Ter várias formas de fazer a mesma coisa (ex: diferentes tipos de filtros de dados) e querer trocar entre elas facilmente.

Solução: Define uma família de algoritmos e os torna intercambiáveis em tempo de execução.

Componentes: Strategy (interface do algoritmo), ConcreteStrategy (a implementação real) e Context (quem usa a estratégia).

Template Method

Problemas: Vários processos seguem os mesmos passos básicos, mas alguns detalhes mudam entre eles.

Solução: Define o esqueleto do algoritmo em uma classe pai, deixando que as classes filhas preencham apenas as partes específicas.

Componentes: AbstractClass (define o molde do método) e ConcreteClass (preenche os buracos do molde).

Visitor

Problemas: Precisar adicionar novas operações a um grupo de classes sem ter que mexer no código de cada uma delas.

Solução: O objeto "aceita" um visitante que entra na classe e executa a lógica externa usando os dados do objeto.

Componentes: Visitor (declara as operações), ConcreteVisitor (a lógica nova) e Element (a interface que aceita o visitante).

Exemplo Prático: Padrão Strategy

Filtragem de base de alunos do setor acadêmico.

```python
from abc import ABC, abstractmethod

# Interface da Estratégia
class FiltroBase(ABC):
    @abstractmethod
    def aplicar(self, dados):
        pass

class FiltroAtivos(FiltroBase):
    def aplicar(self, dados):
        return [d for d in dados if d['status'] == 'Ativo']

class FiltroPeriodo(FiltroBase):
    def __init__(self, periodo):
        self.periodo = periodo
        
    def aplicar(self, dados):
        return [d for d in dados if d['periodo'] == self.periodo]

class ProcessadorRelatorio:
    def __init__(self, filtro: FiltroBase):
        self.filtro = filtro

    def set_filtro(self, filtro: FiltroBase):
        self.filtro = filtro

    def gerar(self, base_dados):
        resultado = self.filtro.aplicar(base_dados)
        print(f"Relatório gerado. Total de registros: {len(resultado)}")
        return resultado

base_cesar = [
    {"nome": "Jonas", "status": "Ativo", "periodo": 1},
    {"nome": "Adriano", "status": "Ativo", "periodo": 2},
    {"nome": "Fulano", "status": "Inativo", "periodo": 1}
]

relatorio = ProcessadorRelatorio(FiltroAtivos())
print("Filtro: Ativos")
relatorio.gerar(base_cesar)

relatorio.set_filtro(FiltroPeriodo(1))
print("\nFiltro: 1º Período")
relatorio.gerar(base_cesar)
```
