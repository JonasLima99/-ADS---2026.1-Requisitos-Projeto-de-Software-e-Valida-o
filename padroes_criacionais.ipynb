Factory Method:

Problemas que resolve: Quando uma classe não pode antecipar a classe de objetos que deve criar, ou quando você quer que as subclasses decidam quais objetos instanciar. Evita o acoplamento direto.

Como resolve: Define uma interface para criar um objeto, mas deixa as subclasses alterarem o tipo de objetos que serão criados.

Abstract Factory:

Problemas que resolve: Quando o sistema precisa ser independente de como seus produtos são criados e você precisa trabalhar com famílias de objetos relacionados (ex: botões e janelas que devem ter o mesmo estilo visual).

Como resolve: Fornece uma interface para criar famílias de objetos sem especificar suas classes concretas.

Builder:

Problemas que resolve: O "problema do construtor telescópico" (quando uma classe tem muitos parâmetros no __init__, tornando-o confuso) e a necessidade de criar diferentes representações de um mesmo objeto complexo.

Como resolve: Separa a construção de um objeto complexo da sua representação, permitindo que o mesmo processo de construção crie diferentes resultados.

Prototype:

Problemas que resolve: Quando o custo de criar um novo objeto do zero é alto ou complexo, ou quando você quer evitar uma hierarquia de fábricas.

Como resolve: Cria novos objetos copiando um exemplar existente (clonagem).


Singleton:

Problemas que resolve: Garante que uma classe tenha apenas uma instância em todo o ciclo de vida da aplicação (ex: conexão com banco de dados ou loggers) e fornece um ponto de acesso global a ela.

Como resolve: Torna o construtor privado e define um método estático que retorna a instância única.

```python
# Singleton .py

class DatabaseConnection:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            print("Configurando nova conexão com o banco de dados do CESAR...")
            cls._instance = super(DatabaseConnection, cls).__new__(cls)
        return cls._instance

# Teste
if __name__ == "__main__":
    db1 = DatabaseConnection()
    db2 = DatabaseConnection()

    print(f"ID db1: {id(db1)}")
    print(f"ID db2: {id(db2)}")

    if db1 is db2:
        print("Sucesso: Ambas as variáveis apontam para a mesma instância.")
    else:
        print("Erro: As instâncias são diferentes.")
```
