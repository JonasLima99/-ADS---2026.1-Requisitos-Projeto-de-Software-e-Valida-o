Atividade 2: Padrões de Projeto Estruturais

Você deverá pesquisar os seguintes padrões: Adapter, Bridge, Composite, Decorator, Facade, Flyweight e Proxy.

Adapter:

Problemas que resolve: Quando classes com interfaces incompatíveis precisam trabalhar juntas. É muito comum quando você integra bibliotecas de terceiros ou sistemas legados ao seu código novo sem querer alterar o código original.

Como resolve: Cria uma classe intermediária (o adaptador) que traduz a interface de uma classe para outra que o cliente espera encontrar.

Bridge:

Problemas que resolve: Evita a "explosão" de subclasses quando você tem variações tanto na interface quanto na implementação. Resolve o problema de ter classes muito grandes que tentam fazer tudo.

Como resolve: Divide uma classe grande em duas hierarquias separadas (Abstração e Implementação) que podem ser desenvolvidas de forma independente, ligando-as via composição.

Composite:

Problemas que resolve: Quando você precisa tratar objetos individuais e grupos de objetos de maneira idêntica, como em uma estrutura de pastas e arquivos ou em menus de sistemas.

Como resolve: Organiza os objetos em estruturas de árvore. Tanto os objetos simples (folhas) quanto os recipientes (compostos) implementam a mesma interface, permitindo que o cliente os trate de forma uniforme.

Decorator:

Problemas que resolve: Quando você precisa adicionar funcionalidades a um objeto em tempo de execução de forma flexível, sem usar herança (que é estática e pode gerar subclasses infinitas).

Como resolve: Coloca o objeto original dentro de objetos "invólucros" (wrappers) que contêm o novo comportamento. Você pode empilhar vários decoradores um sobre o outro.

Facade:

Problemas que resolve: Quando um sistema é muito complexo, com dezenas de classes e dependências, tornando difícil para o cliente saber o que chamar para realizar uma tarefa simples.

Como resolve: Fornece uma interface única e simplificada (uma "fachada") que mascara a complexidade do subsistema, expondo apenas o que é essencial.

Flyweight:

Problemas que resolve: Consumo excessivo de memória RAM ao criar uma quantidade enorme de objetos que compartilham muitos dados idênticos.

Como resolve: Separa os dados em Intrínsecos (fixos e compartilhados) e Extrínsecos (únicos de cada instância), mantendo os dados comuns em um cache para serem reutilizados por vários objetos.

Proxy:

Problemas que resolve: Quando você precisa de um controle de acesso, log, cache ou adiamento de carga (lazy loading) para um objeto que é caro ou sensível para ser acessado diretamente.

Como resolve: Cria um objeto substituto (o Proxy) que tem a mesma interface do objeto real, agindo como um intermediário que decide quando e como repassar as chamadas ao objeto original.

Implementação Prática: Padrão Adapter

Escolhi o Adapter por ser extremamente útil para lidar com a diversidade de dados na secretaria acadêmica. O exemplo abaixo simula o tratamento de dados vindos de um sistema externo com nomes de campos diferentes do nosso padrão interno.

```python
import pandas as pd

class ProcessadorAcademico:
    def gerar_relatorio_ativos(self, df):
        print("Filtro aplicado: Alunos com status 'Ativo'")

        return df[df['status'] == 'Ativo']

class SistemaExternoAPI:
    def obter_dados_brutos(self):

        return [
            {"full_name": "Jonas de Lima Neto", "situation": "Ativo"},
            {"full_name": "Adriele Silva", "situation": "Inativo"},
            {"full_name": "Maria Monteiro", "situation": "Ativo"}
        ]

class AcademicoAdapter:
    def __init__(self, api_externa):
        self.api_externa = api_externa

    def preparar_dataframe(self):
        dados = self.api_externa.obter_dados_brutos()
        df = pd.DataFrame(dados)

        df = df.rename(columns={
            'full_name': 'nome',
            'situation': 'status'
        })
        return df

api_externa = SistemaExternoAPI()
adapter = AcademicoAdapter(api_externa)
processador = ProcessadorAcademico()

df_final = adapter.preparar_dataframe()
relatorio = processador.gerar_relatorio_ativos(df_final)

print("\nRelatório de Alunos Ativos:")
print(relatorio)
```

Ao estudar esses padrões, surge a dúvida: quando parar de criar classes e focar na simplicidade (KISS - Keep It Simple, Stupid)?

Padrões estruturais como o Bridge ou Decorator trazem muita flexibilidade, mas tornam o código mais difícil de rastrear para quem não conhece o projeto. Em scripts de automação simples na secretaria, um if/else direto pode ser mais eficiente que montar toda uma estrutura de padrões.

O Adapter é excelente quando sabemos que fontes de dados mudarão sempre. Porém, se o dado é fixo, criar um Adapter é "overengineering".

O objetivo final de um padrão deve ser facilitar a vida do desenvolvedor a longo prazo. Se o padrão está tornando o código "místico" ou difícil de explicar para um colega de equipe, talvez a solução simples fosse a melhor escolha.
